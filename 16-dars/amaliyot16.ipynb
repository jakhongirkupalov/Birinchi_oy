# ONE HOT ENCODING

```python
import pandas as pd
encoded = pd.get_dummies(df["Drug"], prefix="col", dtype=int)
df=pd.concat([df.drop("Drug", axis = 1), encoded], axis=1)
```

```python
df.info()
```

# LABEL ENCODING

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

df["Status"] = encoder.fit_transform(df["Status"])
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
freq = df["Ascites"].value_counts()
df["Ascites_freq"] = df["Ascites"].map(freq)
```

```python
df.info()
```

# TARGET ENCODING

```python
df["Edema"] = df["Edema"].map({"Yes": 1, "No": 0})
mean_target = df.groupby("Stage")["Edema"].mean()
df["Stage_col"] = df["Edema"].map(mean_target)
```

```python
df.info()
```
