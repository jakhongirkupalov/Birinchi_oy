```python
def tozala():
        for col in df2.columns:
            if df2[col].isnull().any():
                if df2[col].dtype == 'object':
                    df2[col] = df2[col].fillna(df2[col].mode()[0])
                else:
                    df2[col] = df2[col].fillna(df2[col].mean())
        return
```

# Feature Engineering qilish jarayoni

```python
ustunlar = df2.columns.tolist()
print(ustunlar)
```

```python
import pandas as pd
import numpy as np

# Sanani datetime formatga o'tkazish
df2['Date'] = pd.to_datetime(df2['Date'])

# Sanadan chiqqan xususiyatlar
df2['year'] = df2['Date'].dt.year
df2['month'] = df2['Date'].dt.month
df2['day'] = df2['Date'].dt.day
df2['day_of_week'] = df2['Date'].dt.dayofweek  # Dushanba=0, Yakshanba=6

# Returnlar
df2['return_1d'] = df2['close'].pct_change(1)  # 1 kunlik return
df2['return_5d'] = df2['close'].pct_change(5)  # 5 kunlik return

# Moving average'lar (harakatlanuvchi o'rtacha)
df2['ma_5'] = df2['close'].rolling(window=5).mean()
df2['ma_10'] = df2['close'].rolling(window=10).mean()
df2['ma_20'] = df2['close'].rolling(window=20).mean()

# Volatility (yaxshi ko‘rsatkich)
df2['volatility_5'] = df2['close'].rolling(window=5).std()
df2['volatility_10'] = df2['close'].rolling(window=10).std()

# Kunlik narxlar farqlari
df2['price_range'] = df2['high'] - df2['low']
df2['body'] = df2['close'] - df2['open']  # candle body

# Volume o'zgarishi
df2['volume_change_1d'] = df2['volume'].pct_change(1)
df2['volume_rolling_mean_5'] = df2['volume'].rolling(window=5).mean()

# Lag features (oldingi kunlardagi qiymatlar)
df2['close_lag_1'] = df2['close'].shift(1)
df2['close_lag_3'] = df2['close'].shift(3)

# Price ratios
df2['high_low_ratio'] = df2['high'] / df2['low']
df2['close_open_ratio'] = df2['close'] / df2['open']

# Ma'lumotni tozalash (NaN larni yo'qotamiz)
df2 = df2.dropna().reset_index(drop=True)
```
